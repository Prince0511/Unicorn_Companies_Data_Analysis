```python
# Select the investors of interest
investors_of_interest = ['Sequoia Capital China', 'Tiger Global Management', 'Khosla Ventures', 'Andreessen Horowitz', 'index Ventures', 'Sequoia Capital', 'SoftBank Group', 'Hillhouse Capital Management', 'General Catalyst', 'DST Global']
```

```python
# Extract unique investors and create nodes
investors = set()
for entry in data:
    investors.update(entry['Investors'].split(', '))
for investor in investors:
    if investor in investors_of_interest:
        G.add_node(investor)
```

```python
# Create edges between investors and companies with weights representing investment counts
for entry in data:
    company = entry['Company']
    industry = entry['Industry']
    investors = entry['Investors'].split(', ')
    for investor in investors:
        if investor in investors_of_interest:
            if G.has_edge(investor, company):
                G[investor][company]['weight'] += 1
            else:
                G.add_edge(investor, company, weight=1, industry=industry)
```

```python
# Draw the network graph
pos = nx.spring_layout(G, seed=42, k=2)  # Define the layout for the graph
edge_weights = [G[u][v]['weight'] * 0.1 for u, v in G.edges()]  # Scale edge weights for visualization
node_sizes = [500 if node in investors_of_interest else 300 for node in G.nodes()]  # Set larger node size for investors of interest
node_colors = ['pink' if node in investors_of_interest else 'lightgreen' for node in G.nodes()]  # Set different node color for investors of interest
edge_colors = ['black' if G[u][v]['industry'] in ['Artificial intelligence', 'Other', 'E-commerce & direct-to-consumer', 'Fintech', 'Internet software & services', 'Supply chain, logistics, & delivery', 'Consumer & retail', 'Data management & analytics', 'Edtech', 'Health', 'Hardware', 'Auto & transportation', 'Travel', 'Cybersecurity', 'Mobile & telecommunications'] else 'blue' for u, v in G.edges()] # Set different edge color for fintech industry
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=node_colors)  # Draw the nodes
nx.draw_networkx_edges(G, pos, edge_color=edge_colors, alpha=1, width=edge_weights*2, arrows=False)  # Draw the edges with weights as edge width
nx.draw_networkx_labels(G, pos, labels={node: node for node in G.nodes()}, font_size=7, font_color='black')  # Draw the node labels

# Customize the visualization
plt.title('Investor-Company Network Graph')
plt.axis('off')  # Turn off the axis
plt.tight_layout()  # Adjust layout
plt.show()  # Show the graph visualization
```
